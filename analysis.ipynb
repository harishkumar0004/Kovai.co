```python
import pandas as pd
```

```python
df = pd.read_csv("d:/kovai-data-science-intern/data/data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.fillna(0,inplace =True)
df['Total']= df[['Local Route', 'Light Rail','Peak Service','Rapid Route', 'School','Other']].sum(axis=1)
```

```python
df.info()
```

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_summary = df.groupby('Month')[['Local Route', 'Light Rail','Peak Service','Rapid Route', 'School','Other','Total']].sum()
monthly_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
plt.subplot(3,2,1)
plt.plot(df.index, df['Peak Service'])
plt.title("Peakservice")


plt.subplot(3,2,2)
plt.plot(df.index, df['Light Rail'])
plt.title("Light Rail")


plt.subplot(3,2,3)
plt.plot(df.index, df['Rapid Route'])
plt.title("Rapid Route")


plt.subplot(3,2,4)
plt.plot(df.index, df['School'])
plt.title("School")


plt.subplot(3,2,5)
plt.plot(df.index, df['Other'])
plt.title("Other")
```

```python
monthly_summary['Total'].plot(figsize=(16,6),title = 'Totla passengers over time')
plt.ylabel('Passsengers')
plt.show()
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['weekday'] = df['Date'].dt.day_name()
df['year-month'] = df['Date'].dt.to_period('M').astype(str)
```

```python
df.head()
```

```python
bar = df.groupby('year-month')['Total'].sum().reset_index()
```

```python
import seaborn as sns

sns.barplot(data=bar, x= 'year-month', y= 'Total', palette='viridis')

plt.figure(figsize=(16,6))
plt.xlabel('Month')
plt.ylabel('Total passengers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
